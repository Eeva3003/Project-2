```python
df['FastCharge_KmH'] = df['FastCharge_KmH'].fillna(df['FastCharge_KmH'].mean())
```

```python
df = df.dropna(subset=['FastCharge_KmH'])
```

```python
print(df[['Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro']].isnull().sum())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro']])

```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
from sklearn.cluster import KMeans

# KMeans Clustering
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method', fontsize=15)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Selecting the optimal number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_pca)

# Adding cluster labels to the original dataframe
df['Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_kmeans, palette='viridis', s=100, alpha=0.8)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.title('Clusters of Electric Vehicles', fontsize=15)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
